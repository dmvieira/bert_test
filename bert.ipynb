```python
from transformers import BertModel, BertConfig, BertTokenizer, AutoTokenizer, AutoModel
import torch
```

```python
configuration = BertConfig()
model = BertModel(configuration)
```

```python
pre_model = BertModel.from_pretrained('bert-base-uncased')
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python

input_ids = torch.tensor(tokenizer.encode("Hello, my dog is cute", add_special_tokens=True)).unsqueeze(0)  # Batch size 1
outputs = model(input_ids)

last_hidden_states = outputs[0] 
```

# Using API

```python
from bert_serving.client import BertClient
bc = BertClient()
```

```python
rei = bc.encode(["rei"])
```

```python
homem = bc.encode(["homem"])
```

```python
mulher = bc.encode(["mulher"])
```

```python
rainha = bc.encode(["rainha"])
```

```python
rei - homem + mulher
```

```python
rainha
```

```python
from torch.nn import CosineSimilarity
```

```python
cos = CosineSimilarity(dim=1, eps=1e-6)
```

```python
cos(torch.from_numpy(rainha), torch.from_numpy(rei - homem + mulher))
```

```python
cos(torch.from_numpy(rainha), torch.from_numpy(rei))
```
